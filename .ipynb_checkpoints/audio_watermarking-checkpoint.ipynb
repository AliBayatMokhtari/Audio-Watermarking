```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
```

```python
rate, signal = wavfile.read('./ABM_Bye1.wav')

# plotting some informations
plt.title('Speech Signal')
plt.xlabel('Time')
plt.ylabel('Value')
plt.plot(signal)
plt.show()
```

```python
def framing(signal, rate, frame_length):
    frames = []
    frame_size = int(frame_length * rate)
    num_frames = len(signal) // frame_size
    
    for i in range(num_frames):
        frame = signal[i * frame_size : (i + 1) * frame_size]
        frame = np.hamming(len(frame)) * frame
        frames.append(frame)
    
    return np.asarray(frames)
```

```python
signal_frames = framing(signal, rate, frame_length=0.010)
```

```python
def ste(frames):
    energies = []
    for frame in frames:
        energy = np.sum(np.power(frame, 2))
        energies.append(energy)
    
    # plotting some informations
    plt.title('Energy of signal frames')
    plt.xlabel('Frame')
    plt.ylabel('Energy')
    plt.plot(energies)
    plt.show()
    
    return energies
```

```python
energy = ste(signal_frames)
```

```python
def zcc(frames):
    def sign(x):
        return 1 if x >= 0 else -1
    
    zccs = []
    
    def calculate_zcr_frames():
        for frame in frames:
            zc_rate = 0
            for i in range(1, len(frame)):
                zc_rate += 0.5 * abs(sign(frame[i]) - sign(frame[i - 1]))
            zccs.append(zc_rate)
    
    calculate_zcr_frames()
        
    # plotting some informations
    plt.title('Zero Crossing Counts of signal frames')
    plt.xlabel('Frame')
    plt.ylabel('ZCC')
    plt.plot(zccs)
    plt.show()
        
    return np.asarray(zccs)
            
```

```python
zccs = zcc(signal_frames)
```
